## Завдання 5

Створіть об'єкт лінійного регресора, навчіть його на всіх даних і виведіть ваги моделі (ваги зберігаються у змінній `coef_` класу регресора). Можна виводити пари (назва ознаки, вага), скориставшись функцією `zip`, вбудованою в мову python. Назви ознак зберігаються у змінній `df.columns`.


```python
multicol_model = LinearRegression().fit(X, y)
for weight, feature in zip(multicol_model.coef_, df.columns[:-1]):
    print(f"{feature}: {weight}")
```

Ми бачимо, що ваги при лінійно-залежних ознаках за модулем значно більші, ніж при інших ознаках. Щоб зрозуміти, чому так сталося, згадаємо аналітичну формулу, за якою обчислюються ваги лінійної моделі в методі найменших квадратів:

$$w = (X^TX)^{-1} X^T y$$

Якщо в $X$ є колінеарні (лінійно-залежні) стовпці, матриця $X^TX$ стає виродженою, і формула перестає бути коректною. Чим більш залежні ознаки, тим менший визначник цієї матриці й тим гірша апроксимація $Xw \approx y$. Таку ситуацію називають _проблемою мультиколінеарності_.

З парою (temp, atemp) трохи менше корелюючих змінних такого не сталося, однак на практиці завжди варто уважно стежити за коефіцієнтами при схожих ознаках.

Для того щоб розв'язати проблему мультиколінеарності, потрібно скористатися регуляризатором. До оптимізуючого функціоналу додають $L_1$ або $L_2$ норму ваг, помножену на коефіцієнт регуляризації $\alpha$. У першому випадку метод називається Lasso, а у другому — Ridge.

### Завдання 6
Давайте спробуємо навчити лінійні моделі з $L_1$ і $L_2$-регуляризацією, а далі порівняємо їхні ваги. Навчіть регресори Ridge і Lasso з параметрами за замовчуванням і переконайтеся, що проблема з вагами вирішилась.

Навчіть лінійну модель з $L_1$-регуляризацією (клас Lasso) і виведіть ваги.

```python
from sklearn.linear_model import Lasso, Ridge
lasso_model = Lasso().fit(X, y)
for weight, feature in zip(lasso_model.coef_, df.columns[:-1]):
    print(f"{feature}: {weight}")
```

Навчіть лінійну модель з $L_2$-регуляризацією (клас Ridge) і виведіть ваги.

```python
ridge_model = Ridge().fit(X, y)
for weight, feature in zip(ridge_model.coef_, df.columns[:-1]):
    print(f"{feature}: {weight}")
```

### Завдання 7

На відміну від $L_2$-регуляризації, $L_1$ обнуляє ваги при деяких ознаках. Давайте поспостерігаємо, як змінюються ваги зі збільшенням коефіцієнта регуляризації $\alpha$.

Для кожного значення коефіцієнта з `alphas` навчіть регресор `Lasso` і запишіть ваги у список `coefs_lasso`, а потім навчіть `Ridge` і запишіть ваги у список`coefs_ridge`. Конвертуйте отримані списки в `np.array`.

```python
alphas = np.arange(1, 500, 50)

coefs_lasso = []
coefs_ridge = []

for alpha in alphas:
    lasso = Lasso(alpha=alpha).fit(X, y)
    ridge = Ridge(alpha=alpha).fit(X, y)
    coefs_lasso.append(lasso.coef_)
    coefs_ridge.append(ridge.coef_)

coefs_lasso = np.array(coefs_lasso)
coefs_ridge = np.array(coefs_ridge)
```

Проаналізуйте динаміку ваг при збільшенні параметра регуляризації:

```python
plt.figure(figsize=(8, 5))

for coef, feature in zip(coefs_lasso.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))

plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Lasso")

plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_ridge.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))

plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Ridge")

plt.show()
```

__Відповідь:__

Регуляризатор Lasso більш агресивно зменшує ваги при одному й тому самому alpha.

Якщо alpha зробити дуже великим, то в вагах lasso почнуть утворюватись нулі в більшості ознак. Так відбувається через те що, чим більше значення alpha, тим більше штраф за величину ваг.

Ні, не можна стверджувати, що `lasso` виключає одну з ознак windspeed при будь якому значенні `alpha > 0`, тому що alpha може виключити ознаку, якщо її коефіцієнт стане меншим за деякий поріг. Якщо ознака windspeed має малий вплив на цільову змінну, то при певному значенні alpha Lasso може обнулити її коефіцієнт, тим самим виключивши її з моделі.
А Ridge не виключає ознаки, а лише зменшує величину їхніх коефіцієнтів. Ridge працює так, що при великому alpha коефіцієнти можуть стати дуже малими, але не стануть рівними нулю.

Для відбору неінформативних ознак краще підходить Lasso, так як цей регуляризатор може виключити неінформативну ознаку, що обнулить коефіцієнт впливу ознаки на ваги.

### Завдання 8

Далі будемо працювати з `Lasso`.

Отже, ми бачимо, що при зміні alpha модель по-різному підбирає коефіцієнти ознак. Нам потрібно вибрати найкраще alpha.

Для цього, по-перше, нам потрібна метрика якості. Будемо використовувати як метрику сам оптимізований функціонал методу найменших квадратів, тобто `Mean Square Error`.

По-друге, потрібно зрозуміти, на яких даних цю метрику рахувати. Не можна вибирати `alpha` за значенням MSE на навчальній вибірці, тому що тоді ми не зможемо оцінити, як модель буде робити передбачення на нових для неї даних. Якщо ми виберемо одне розбиття вибірки на навчальну та тестову (це називається holdout), то налаштуємося на конкретні "нові" дані, і знову можемо перенавчитися. Тому будемо робити декілька розбиттів вибірки, на кожному пробувати різні значення alpha, а потім усереднювати MSE. Найзручніше робити такі розбиття крос-валідацією, тобто розділити вибірку на $K$ частин, і кожного разу брати одну з них як тестову, а з блоків, що залишилися, складати навчальну вибірку.

Робити крос-валідацію для регресії в sklearn зовсім просто: для цього є спеціальний регресор, __LassoCV__, який бере на вхід список із alpha і для кожного з них обчислює MSE на крос-валідації. Після навчання (якщо залишити параметр cv=3 за замовчуванням) регресор буде містити змінну __mse\_path\___, матрицю розміру len(alpha) x k, k = 3 (число блоків у крос-валідації), що містить значення MSE на тесті для відповідних запусків. Крім того, у змінній alpha\_ буде зберігатися вибране значення параметра регуляризації, а в coef\_, традиційно, навчені ваги, що відповідають цьому alpha_.

Зверніть увагу, що регресор може змінювати порядок, у якому він проходить по alphas; для зіставлення з матрицею MSE краще використовувати змінну регресора alphas_.

Навчіть регресор `LassoCV` на всіх параметрах регуляризації з alpha. Побудуйте графік _усередненого_ за рядками `MSE` в залежності від `alpha` (використовуйте для цього функцію `create_plot`).

Виведіть вибране `alpha`, а також пари "ознака-коефіцієнт" для навченого вектора коефіцієнтів.

```python
def create_plot(data, title, xlabel, ylabel, figsize=None):
    size = figsize or (15, 5)
    plt.figure(figsize=size)

    x = list(map(lambda e: e[0], data))
    y = list(map(lambda e: e[1], data))
    plt.plot(x, y)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.grid()
    plt.show()
```

```python
from sklearn.linear_model import LassoCV

alphas = np.arange(1, 100, 5)

regressor = LassoCV(alphas=alphas, cv=3).fit(X, y)

mse_data = list(zip(regressor.alphas_, np.mean(regressor.mse_path_, axis=1)))

create_plot(mse_data, 'Mean Squared Error vs Alpha', 'Alpha', 'Mean Squared Error')

print(f"Вибране alpha: {regressor.alpha_}")

coef_pairs = zip(df.columns[:-1], regressor.coef_)
for feature, coef in coef_pairs:
    print(f"{feature}: {coef}")
```

Отже, ми вибрали певний параметр регуляризації. Давайте подивимося, які б ми вибирали alpha, якби ділили вибірку лише один раз на навчальну та тестову, тобто розглянемо траєкторії MSE, що відповідають окремим блокам вибірки.

### Завдання 9

Виведіть значення `alpha`, що відповідають мінімумам `MSE` на кожному розбитті (тобто за стовпцями).
На трьох окремих графіках візуалізуйте стовпці `mse_path_`.


```python
print("Alphas with minimum MSE:")
for i in range(regressor.mse_path_.shape[1]):
    min_mse_idx = np.argmin(regressor.mse_path_[:, i])
    best_alpha = regressor.alphas_[min_mse_idx]
    best_mse = regressor.mse_path_[min_mse_idx, i]
    print(f"Alpha = {best_alpha}, MSE = {best_mse}")

for i in range(regressor.mse_path_.shape[1]):
    plt.figure(figsize=(15, 5))
    plt.plot(regressor.alphas_, regressor.mse_path_[:, i])
    plt.title(f'MSE для крос-валідаційного розбиття {i + 1}')
    plt.xlabel('Alpha')
    plt.ylabel('MSE')
    plt.grid()
    plt.show()
```

На кожному розбитті оптимальне значення `alpha` своє, і йому відповідає велике MSE на інших розбиттях. Виходить, що ми налаштовуємося на конкретні навчальні та контрольні вибірки. При виборі `alpha` на крос-валідації ми вибираємо дещо "середнє", що даватиме прийнятне значення метрики на різних розбиттях вибірки.

Чи видно зростаючу лінійну залежність `cnt` від цих ознак за графіками? Чи логічно стверджувати (виходячи зі здорового глузду), що чим більше значення цих ознак, тим більше людей захочуть взяти велосипеди?

__Відповідь:__

season: 532.0189828413535

yr: 1015.0602226430597

temp: 370.6798550300328

atemp: 581.3969310654992

З вибраними ознаками, такими як season, yr, temp, і atemp, ми бачимо позитивні коефіцієнти, що вказує на те, що ці фактори позитивно корелюють з кількістю орендованих велосипедів. Однак на графіках може бути помітно, що для деяких ознак, таких як температура або відчутна температура, зростання значень на початку дійсно призводить до збільшення кількості орендованих велосипедів, але після певної точки зростання значень ці ознаки починають мати зворотний ефект — кількість орендованих велосипедів починає знижуватись.

Це може означати, що певні умови, наприклад, дуже висока температура або температура на рівні жаркого літа, можуть призводити до того, що люди насправді менш охоче орендують велосипеди, оскільки це може бути незручно або навіть небезпечно. Влітку можуть бути деякі піки в орендах при комфортних температурах, але екстремальні значення температури можуть мати протилежний ефект.

Тому висновок, що "чим більше значення цих ознак, тим більше людей захочуть взяти велосипеди", не завжди є логічним. Існує оптимальний діапазон значень, за межами якого інтенсивність оренд зменшується.

#### Запитання 2

Виберіть 3 ознаки з найбільшими за модулем негативними коефіцієнтами (і випишіть їх), подивіться на відповідні візуалізації. Чи видно лінійну залежність, що зменшується? Чи логічно стверджувати, що чим більша величина цих ознак, тим менше людей захочуть взяти велосипеди?

__Відповідь:__

weathersit: -330.5598567399813

hum: -140.00740550068875

windspeed(mph): -191.7714084713494

Ми спостерігаємо негативні коефіцієнти, що вказує на зворотну кореляцію цих ознак з кількістю орендованих велосипедів. Це означає, що збільшення значень цих факторів призводить до зменшення кількості оренд.

На графіках залежностей видно, що зі збільшенням значень цих ознак дійсно зменшується кількість орендованих велосипедів, що підтверджує лінійну залежність.

Висновок: Логічно стверджувати, що чим більша величина цих ознак, тим менше людей захочуть орендувати велосипеди. Це може бути обумовлено такими факторами:

- weathersit (ситуація з погодою): Погана погода (дощ, туман, злива) може негативно впливати на бажання людей брати велосипеди.
- hum (вологість): Висока вологість може робити умови для їзди на велосипеді неприємними.
- windspeed(mph) (швидкість вітру): Сильний вітер також може створювати незручності та небезпеки для велосипедистів.

#### Запитання 3

Випишіть ознаки з коефіцієнтами, близькими до нуля (< 1e-3). Як ви думаєте, чому модель виключила їх із моделі (знову подивіться на графіки)? Чи правда, що вони ніяк не впливають на попит на велосипеди?

__Відповідь:__

windspeed(ms): -2.685570636704926e-08

Ознака, яка має коефіцієнт близький до нуля, це windspeed(ms) з коефіцієнтом -2.69e-08. Це вказує на те, що модель не вважає її важливою для прогнозування попиту на велосипеди.

Чому модель виключила цю ознаку?

Це може бути пов'язано з сильною кореляцією між windspeed(ms) та windspeed(mph). Оскільки windspeed(ms) та windspeed(mph) дуже схожі за значенням (лише мають різні одиниці вимірювання), модель, ймовірно, виявила, що ця ознака не додає нової інформації для прогнозу. Зазвичай LassoCV намагається зменшити коефіцієнти для подібних ознак, які мають майже однакову інформацію, і, як результат, одна з них (у даному випадку windspeed(ms)) стає дуже малою.

Чи правда, що ця ознака не впливає на попит?

Оскільки коефіцієнт дуже близький до нуля, це означає, що модель не вважає windspeed(ms) важливим фактором для прогнозування попиту на велосипеди. Можливо, її роль вже повністю виконує windspeed(mph), і додавання цієї ознаки до моделі не дає значного внеску в точність прогнозу.

Висновок:

Ознака windspeed(ms) має дуже малий коефіцієнт через високу кореляцію з windspeed(mph). Модель, ймовірно, вважає, що ця ознака не додає додаткової цінної інформації для прогнозу попиту на велосипеди, і тому її коефіцієнт наближається до нуля.

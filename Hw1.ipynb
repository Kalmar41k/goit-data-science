Створіть наступний шаблон без жорсткого кодування(Hard coding) . Використовуйте лише функції Numpy і наведений нижче масив введення Input:a = np.array([1,2,3])

Output:array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])

```python
a = np.array([1, 2, 3])
repeat_count = len(a)
output = np.repeat(a, repeat_count)
output = np.concatenate([output, np.tile(a, np.shape(a)).flatten()])
print(output)
```

Знайдіть повторювані записи (від 2-го випадку повторення і далі) у заданому масиві і позначте їх як True. Перше повторення має бути False.

\# Input np.random.seed(100) a = np.random.randint(0, 5, 10)

[0 0 3 0 2 4 2 2 2 2] #Output [False True False True False False True True True True]

```python
np.random.seed(100)

a = np.random.randint(0, 5, 10)

unique, counts = np.unique(a, return_counts=True)

output = np.zeros_like(a, dtype=bool)

for value, count in zip(unique, counts):
    if count > 1:
        indices = np.where(a == value)[0]
        output[indices[1:]] = True

print(output)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

 # Завантажуємо таблицю

```python
io = "http://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
data = pd.read_html(io, attrs={"class": "wikitable collapsible collapsed"}, decimal=",", thousands=".")
df = data[3]
df
```

 # Вивести перші рядки таблиці за допомогою методу head

```python
df.head()
```

 # Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

 # Замініть у таблиці значення "—" на значення NaN

```python
df1 = df.replace("—", np.nan)
df1
```

 # Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df1.dtypes
```

 # Замініть типи нечислових колонок на числові.

```python
df1["2014"] = df1["2014"].astype(float)
df1["2019"] = df1["2019"].astype(float)
df1.dtypes
```

 # Порахуйте, яка частка пропусків міститься в кожній колонці 

```python
df1.isnull().sum() / len(df1) * 100
```

 # Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df1 = df1.drop(df1.index[-1])
df1
```

 # Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python

df1.iloc[:, 1:] = df1.iloc[:, 1:].fillna(df1.iloc[:, 1:].mean())
df1 = df1.round(1)
df1
```

 # Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df1["Регіон"][df1["2019"] > df1["2019"].mean()]
```

 # У якому регіоні була найвища народжуваність у 2014 році?

```python
df1.loc[df1["2014"].idxmax()]["Регіон"]
```

 # Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))
sns.barplot(x="2019", y="Регіон", data=df1, hue="Регіон")
plt.xlabel("Коефіцієнт народжуваності")
plt.ylabel("Регіон")
```

## Візуалізація

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="2014", y="Регіон", data=df1)
plt.xlabel("Коефіцієнт народжуваності")
plt.ylabel("Регіон")
plt.show()
```

```python
plt.figure(figsize=(15, 20))
fig, axs = plt.subplots(2, 1, figsize=(6, 12))
axs[0].barh(df1["Регіон"], df1["1950"], label="1950")
axs[1].barh(df1["Регіон"], df1["1960"], label="1960")
plt.xlabel("Коефіцієнт народжуваності")
plt.ylabel("Регіон")
plt.show()
```
